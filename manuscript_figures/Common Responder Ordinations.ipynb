```python
%%R
physeq
```

```python
%%R
SampleNames = colnames(otu_table(physeq))
MinRoot = c("Root","Root","Root","Root","Mineral","Root","Mineral","Root","Mineral","Root","Mineral","Root")
SampleData = join(data.frame(SampleNames),data.frame(MinRoot))
SampleData$MinRoot = MinRoot
SampleData$Week = c(12,9,6,3,14,12,14,9,14,6,14,3)
SampleData$Season = c(2,2,2,2,1,1,1,1,1,1,1,1)
row.names(SampleData)=SampleData$SampleNames
SampleData$SampleNames = plyr::revalue(SampleData$SampleNames, c("Ferrihydrite"="Ferrihydrite","Quartz"="Quartz","Kaolinite"="Kaolinite","Heavy"="Heavy Fraction"))
SampleData
#Generating the sample data for the physeq table
```

```python
%%R
#colnames(otu_table(physeq))==rownames(SampleData)
sample_data(physeq) = sample_data(SampleData)
physeq
```

```python
%%R
# Computing the Bray distance matrix and NMDS in parallel

ord = ordinate(physeq, method = "NMDS", distance = "jaccard", weighted=TRUE, k=3, trymax=100)
# Consider trymax= and k=3
```

```python
%%R -h 500 -w 600
D <- ord
# Sets the relevant ordination

# Creates an plot object, using the df.sites data, with the x and y axes specified as the PCoA axes,
p = plot_ordination(physeq, D, shape = "MinRoot", color = "Week")

# Determines the size of the symbols
p = p  + geom_point(size=3.5) 

p = p + geom_text(aes(label=SampleNames),nudge_x=0.03, nudge_y=0.02)

p = p + guides(colour = guide_legend(""), shape = guide_legend(""))
p = p + scale_colour_gradient(low="orange",high="red")
#p = p + scale_shape_manual(values=c(19,4,17,15,3))

p = p + scale_x_continuous(breaks=c(-4,-3,-2,-1,0,1))
p = p + scale_y_continuous(breaks=c(-1,0,1))

p = p + theme_bw() +
  theme(
      legend.position = "right",
      legend.justification=c(0,1),
      legend.position=c(0,1),
      legend.box.just= "left",
    plot.background = element_blank()
   ,panel.grid.major = element_blank()
   ,panel.grid.minor = element_blank()
    ,strip.text.x = element_text(size=14, face="bold"),
        axis.title = element_text(size=22, face="bold"),
        axis.text = element_text(size=16),
        legend.text = element_text(size=16),
        legend.title = element_text(size = 22),
        strip.background = element_rect(colour="white", fill="white"))

p
```
